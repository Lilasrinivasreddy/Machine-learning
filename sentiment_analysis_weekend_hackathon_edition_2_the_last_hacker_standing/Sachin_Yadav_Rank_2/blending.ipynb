```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.metrics import log_loss
```

```python
Train_roberta_pro = pd.read_csv('../input/blendrobertampnet/Train_prob_Rb_cased.csv')
Train_mpnet_pro = pd.read_csv('../input/blendrobertampnet/Train_prob_mpnet_cased.csv')

Train_roberta_pro_un = pd.read_csv('../input/uncased/Train_prob_RB_uncased.csv')
Train_mpnet_pro_un = pd.read_csv('../input/uncased/Train_prob_MPNET_uncased.csv')


Test_roberta_pro = pd.read_csv('../input/blendrobertampnet/Test_prob_Rb_cased.csv')
Test_mpnet_pro = pd.read_csv('../input/blendrobertampnet/Test_prob_mpnet_cased.csv')

Test_roberta_pro_un = pd.read_csv('../input/uncased/Test_prob_RB_uncased.csv')
Test_mpnet_pro_un = pd.read_csv('../input/uncased/Test_prob_MPNET_uncased.csv')
```

```python
Target = Train_roberta_pro.pop('Sentiment');
Train_mpnet_pro.pop('Sentiment');

Train_roberta_pro_un.pop('Sentiment');
Train_mpnet_pro_un.pop('Sentiment');
```

```python
Train_roberta_pro = np.array(Train_roberta_pro) 
Train_mpnet_pro =  np.array(Train_mpnet_pro)
Train_roberta_pro_un = np.array(Train_roberta_pro_un) 
Train_mpnet_pro_un =  np.array(Train_mpnet_pro_un)

Test_roberta_pro = np.array(Test_roberta_pro)
Test_mpnet_pro = np.array(Test_mpnet_pro)
Test_roberta_pro_un = np.array(Test_roberta_pro_un)
Test_mpnet_pro_un = np.array(Test_mpnet_pro_un)
```

```python
ensemble_oof = 0.75*Train_roberta_pro + (1-.75)*Train_roberta_pro_un

# Blending of MPNET cased and Uncased was dropped , as it didnt result in good results. 
#ensemble_oof_un = 0.95*Train_mpnet_pro + (1-.95)*Train_mpnet_pro_un

ensemble_oof =  0.4*ensemble_oof + (1-0.4)*Train_mpnet_pro

score = log_loss(Target,ensemble_oof)
score
```

```python
test_ensemble_oof = 0.75*Test_roberta_pro + (1-.75)*Test_roberta_pro_un

final_sub =  0.4*test_ensemble_oof + (1-0.4)*Test_mpnet_pro

final_sub = pd.DataFrame(final_sub)
```

```python
final_sub.to_csv('final_sub_blend99.csv',index=False)
```
