# Start coding -- Load data, then train with AutoGluon

```python
from autogluon import TabularPrediction as task
import pandas as pd

### Change the file location to point to the raw training set
train = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Train.csv')

train_data = task.Dataset(train)

predictor = task.fit(train_data=train_data,
                     label='Class',
                     problem_type='binary', 
                     eval_metric='roc_auc',
                     num_bagging_folds=5,
                     auto_stack=True,
                     num_bagging_sets=5,
                     random_seed=0)
```

```python
### Change the file location to point to the raw test set
test = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Test.csv')

test_data = task.Dataset(test)

sub_df = pd.DataFrame(predictor.predict_proba(test_data, as_pandas=True), columns=['Class'])
sub_df
```

```python
sub_df.to_csv('/content/drive/My Drive/Colab Notebooks/submission_01.csv', index=False)
```
